```python
from sqlite3 import connect
from requests import get
from bs4 import BeautifulSoup
from os import makedirs, path
from matplotlib import pyplot as plt
from tqdm import tqdm_notebook

%matplotlib inline
```

```python
all_names = soup.find_all('h2',{'class': 'zagolovki'})
all_names = [i.text for i in all_names]
all_names
```

```python
def prepeare_move(soup):
    description = {}
    content = soup.find('div', {'class':'fullimg'}).text.strip()
    content = [i for i in content.split('\n') if i]
    description['text'] = content[0]
    
    for elem in content[1:]:
        try:
            item = elem.split(':')
            description[item[0]] = item[1]
        except IndexError:
            pass
    return description
```

```python
all_names_in_10_pages = []

for page in range(10):
    url = f"https://kinogo.by/page/{page}/"
    response = get(url)
    soup = BeautifulSoup(response.text)
    all_names_in_one_page = soup.find_all('h2',{'class': 'zagolovki'})
    all_names_in_one_page = [i.text for i in all_names_in_one_page]
    all_names_in_10_pages.extend(all_names_in_one_page)
```

```python
all_discriptions_in_10_pages = []

for page in tqdm_notebook(range(10), desc="pages"):
    url = f"https://kinogo.by/page/{page}/"
    response = get(url)
    soup = BeautifulSoup(response.text)
    all_span = soup.find_all('span', {'class':'podrobnee'})
    all_urls = [i.find('a').get('href') for i in all_span]
    for url in tqdm_notebook(all_urls, leave=False, desc="move"):
        response = get(url)
        soup = BeautifulSoup(response.text)
        all_discriptions_in_10_pages.append(prepeare_move(soup))
```

```python
with connect("kinogo.db") as conn:
    for name, item in zip(all_names_in_10_pages, all_discriptions_in_10_pages):
        conn.execute("""INSERT INTO movies 
                    values(?,?,?,?,?,?,?,?,?,?,?,?,?)""", 
                     (name,
                    item.get("text"),
                    item.get('Год выпуска'),
                    item.get('Страна'),
                    item.get('Жанр'),
                    item.get('Качество'),
                    item.get('Перевод'),
                    item.get('Продолжительность'),
                    item.get('Кинопоиск'),
                    item.get('IMDB'),
                    item.get('Премьера'),
                    item.get('Режиссер'),
                    item.get('В ролях')))
```

```python
from abc import ABC, abstractmethod
```

```python
class AbstractFigure(ABC):
    @abstractmethod
    def drow(self):
        pass
```

```python
class Rectangle(AbstractFigure):
    def __init__(self, x1, y1, x2, y2):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2
        
    def drow(self):
        
        all_x = list(range(self.x1, self.x2 + 1))
        
        all_y = list(range(self.y1, self.y2 + 1))
      
            
        output =  [(self.x1, i) for i in all_y]
        output += [(i, self.y1) for i in all_x]
        output += [(self.x2, i) for i in all_y]
        output += [(i, self.y2) for i in all_x]
        output = set(output)
        return output
        

class Line(AbstractFigure):
    pass

class BucketFill(AbstractFigure):
    pass
```

```python
class Canvas(AbstractFigure):
    pass
```

```python
rec = Rectangle(1, 1, 3, 4)
field = [['0' for i in range(6)] for i in range(6)]
```

```python
for point in rec.drow():
    field[point[1]][point[0]] = "*"
```
